```python
import pandas as pd
import numpy as np
```

### Ratings

```python
ratings = pd.read_csv(
    "../data/ml-1m/ratings.dat",
    sep="::",
    header=None,
    names=["user_id", "movie_id", "rating", "timestamp"],
    dtype={
        "user_id": np.int32,
        "movie_id": np.int32,
        "ratings": np.float32,
        "timestamp": np.int64,
    },
    engine="python",
    encoding="ISO-8859-1",
)
ratings["timestamp"] = pd.to_datetime(ratings["timestamp"], unit="s")

ratings.head(10)
```

### Movies

```python
movies = pd.read_csv(
    "../data/ml-1m/movies.dat",
    sep="::",
    header=None,
    names=["movie_id", "title", "genre"],
    dtype={
        "movie_id": np.int32,
        "title": str,
        "genre": str,
    },
    engine="python",
    encoding="ISO-8859-1",
)

movies.head(10)
```

### Users

```python
users = pd.read_csv(
    "../data/ml-1m/users.dat",
    sep="::",
    header=None,
    names=["user_id", "gender", "age", "occupation", "zipcode"],
    dtype={
        "user_id": np.int32,
        "gender": str,
        "age": "category",
        "occupation": "category",
        "zipcode": str,
    },
    engine="python",
    encoding="ISO-8859-1",
)

users.head(10)
```
